```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from csv import reader
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df_raw = pd.read_csv("../Dataset/groceries3.csv")
df_raw.head()
```

```python
purchases = []
with open("../Dataset/groceries3.csv", 'r') as read_obj: # 'r' stands for read mode
    csv_reader = reader(read_obj)
    for row in csv_reader:
        purchases.append(row)
```

```python
encoder = TransactionEncoder()
transactions = encoder.fit(purchases).transform(purchases) #.fit finds unique items in pruchases and identifies them. .transform converts this information into a binary form of true and false.
```

```python
transactions = transactions.astype('int')
```

```python
df = pd.DataFrame(transactions, columns=encoder.columns_) # up till cell 7, transactions was a list of listscontaining true and false values for each purchase. Now it is to be converted into a binary matrix data frame.
```

```python
df.head()
```

```python
df = df.loc[:, ~df.columns.str.startswith("Item")]
```

```python
df.head(10)
```

```python
df.shape
```

```python
df = df.loc[:, df.columns.str.strip() != ""] # a safe measure to remove empty sets of items that appeared preiously in the apriori table
```

```python
# applying the apriori algorithm using 2% support
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

frequent_itemsets
```

```python
# sorting the dataframe
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
# finding top 10 items with minimum support of 5%
frequent_itemsets[ (frequent_itemsets['length'] == 1) &
                   (frequent_itemsets['support'] >= 0.05) ][0:10]
```

```python
frequent_itemsets[(frequent_itemsets['length'] > 1) & 
                  (frequent_itemsets['support'] >= 0.01)]
```

```python
rules = association_rules(frequent_itemsets, metric='support', min_threshold=0.02) #MLxtend added zhangs_metric by default and it is not a problem!
rules
```

```python
rules.sort_values(by='confidence', ascending=False)[0:10]
```

```python
rules[(rules['support'] >= 0.02) &
      (rules['lift'] > 1.0)]
```
